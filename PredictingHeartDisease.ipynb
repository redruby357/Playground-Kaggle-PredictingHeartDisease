```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.preprocessing 
from sklearn.preprocessing import StandardScaler 
import seaborn as sns
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import classification_report
```

```python
train_dataset = pd.read_csv('train.csv')
train_dataset.head()
```

```python
test_dataset = pd.read_csv('test.csv')
test_dataset.head()
```

```python
train_dataset.info()
```

```python
train_dataset.describe()
```

```python
features = ['id','Age','Sex','Chest pain type', 'BP','Cholesterol','FBS over 120', 'EKG results', 'Max HR','Exercise angina', 'ST depression', 'Slope of ST', 'Number of vessels fluro', 'Thallium']
y = train_dataset['Heart Disease'].map({'Presence': 1, 'Absence': 0})
```

```python
X = train_dataset[features].copy()
X_True = test_dataset[features].copy()
```

Missing data

```python
print(X.isnull().sum())
```

```python
print(X_True.isna().sum())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_True_scaled = scaler.fit_transform(X_True)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=200)
```

```python
model = xgb.XGBClassifier()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python

print("\nUpdated Classification Report:")
print(classification_report(y_test, pred))
```

```python
model.fit(X_scaled, y)
test_preds = model.predict(X_True_scaled)

```

```python
output = pd.DataFrame({'id': test_dataset.id, 
                       'Heart Disease': test_preds})
output.to_csv(r'C:\Kaggle\submission.csv', index=False)
```
