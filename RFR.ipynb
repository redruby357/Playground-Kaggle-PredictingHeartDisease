```python
label = LabelEncoder()
train_dataset['Heart Disease'] = label.fit_transform(train_dataset['Heart Disease'])
```

```python
train_dataset.head(20)
```

```python
features = ['id','Age','Sex','Chest pain type','BP','Cholesterol','FBS over 120','EKG results','Max HR','Exercise angina','ST depression','Slope of ST','Number of vessels fluro','Thallium']
```

```python
X_train_dataset = train_dataset[features].copy()
X_test_dataset = test_dataset[features].copy()
y_train_dataset = train_dataset['Heart Disease'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_dataset,y_train_dataset,train_size=0.8, random_state= 42)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
print("Results:\n", classification_report(y_test, pred))
```

```python
y_probs  = model.predict_proba(X_test)[:,1]
auc = roc_auc_score(y_test, y_probs)

print("AUC:", auc)
```
